```
# 함수
# 특정 작업을 수행하는 명령어들의 모음에 이름을 붙인 것
# def 함수이름(매개변수1,매개변수2):

# 두 정수를 입력 받아 두 정수의 합계를 반환하는 plus() 만들기
def plus(v1,v2): # 함수 정의
  result=0
  result=v1+v2
  return result
hap=0 # 변수 선언
# 메인 코드
hap=plus(100,200)
print('100과 200의 plus() 함수 결과는 %d'%hap)
```

```
# 누적 함수 sum()
def get_sum(start,end):
  sum=0
  for i in range(start,end+1):
    sum+=i
  return sum
print(get_sum(1,10))
print(get_sum(1,20))

# 55만 돌려주는 함수 sum55()
def sum55():
  sum=0
  for i in range(1,11):
    sum+=i
  return sum
print(sum55())
```

```
# 두 개의 정수를 입력 받아 더 큰 수를 찾아서 반환
def big(a,b):
  return max(a,b)
x,y=map(int,input('정수 2개 입력:').split(' '))
print('큰 수:',big(x,y))
```

```
# 두 개의 정수를 입력 받아 더 큰 수를 찾아서 반환
def big(a,b):
  if a>=b:
    c=a
  else:
    c=b
  return c
x,y=map(int,input('정수 2개 입력:').split(' '))
print('큰 수:',big(x,y))
```

```
# 인수와 매개변수
# 인수:호출 프로그램에 의하여 함수에 실제로 전달되는 값(argument)
# 매개변수:값을 전달받는 변수(parameter)
# 디폴트 인수 def get_sum(start,end=10) 함수의 매개변수가 가지는 기본값
# 반환값:return(함수의 값을 반환)
# 지역변수:한정된 지역에서만 사용되는 변수
# 전역변수:프로그램 전체에서 사용되는 변수
# 지역변수가 있는 함수 안에서 동일한 이름의 전역변수는 사용되지 않는다.
# 전역변수를 사용하는 방법은 global
def func2():
  global a # 바깥쪽에 있는 a를 이제부터 사용합니다.
  print(a)
  a=10
  print(a)
a=20
func2()
print(a)
```

```
# 초를 입력하면 몇 시 몇 분 몇 초인지를 출력하는 calc_time() 함수 작성
def clac_time(sec):
  hour,sec1=divmod(sec,3600)
  min,sec2=divmod(sec1,60)
  return hour,min,sec2
time=int(input('초 입력:'))
if time>=24*3600:
  print('입력이 하루를 초과합니다.')
else:
  h,m,s=clac_time(time)
  print('{0} 시 {1} 분 {2} 초'.format(h,m,s))
```

```
# 카페 무인 주문 함수
# 입력받은 고객 이름과 주문 음료 종류를 출력하는 order()함수를 작성해보자. 단, 고객 이름을 작성하지 않으면 1, 2, ... 등 숫자를 순서대로 출력하고 음료 종류에 '마감'을 입력하면 주문을 종료한다.

def order(name,drink):
  print(name,"고객님, 주문하신",drink,'나왔습니다.')
while True:
  d=input('음료 종류 입력:')
  i=0
  if d=='마감':
    break
  else:
    n=input('고객명 입력:')
    if n=='':
      i+=1
      order(i,d)
    else:
      order(n,d)
```

```
def fact(n):
  if n<=1:
    return 1
  return n*fact(n-1)
print(fact(5))
```

```
# 피보나치 수열
def fact(n):
  if n<=2:
    return 1
  return fact(n-1)+fact(n-2)
print(fact(5))
```
